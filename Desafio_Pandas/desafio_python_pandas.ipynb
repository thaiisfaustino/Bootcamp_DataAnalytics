## **Desafio 6 - Rendimento Ações**

**Definição do problema:** Você está trabalhando num banco e preciso informar ao seu cliente em que ano foi a maior rentabilidade de dividendo pago por ação da empresa Johnson&Jonson

1. Agregar os rendimentos de um mesmo ano

2. Calcular a valorização do rendimento de um ano para outro

3. Informar o ano de maior rentabilidade

**Código do dataset**: JohnsonJohnson

**Dicas:**

- Criar uma nova representação dos dados, onde a informação é anual

- Criar uma nova coluna informando a valorização anual

- numpy.trunc()

- método .shift()

- método .idxmax()

```python
#Carrega o dataframe 'JohnsonJohnson' e atribui à variável df_jj
#Apresentação de apenas as primeiras 10 linhas

df_jj = data('JohnsonJohnson')
df_jj.head(10)
```

```python
#Altera os números da coluna 'time', transformando em inteiro com a retirada dos valores após o ponto

df_jj['time'] = df_jj.apply(lambda row : math.trunc(row['time']),axis=1)
df_jj.head(10)
```

```python
#1. Agregar os rendimentos de um mesmo ano

#Agrupa os valores de 'time' e soma os valores de rendimento ('JohnsonJohnson')

df_jj_agg = df_jj.groupby(['time']).sum()
df_jj_agg
```

```python
#Cria uma nova coluna chamada 'shift'. 
#Ela é uma cópia da coluna 'JohnsonJohnson', que é a única disponível, já que a coluna 'time' se tornou index após o groupby.
#Com o .shift(1), a coluna "move" o valor das linhas para a próxima linha. 
#É como se a linha tivesse em um Id-ano específico, o rendimento daquele ano ('JohnsonJohnson') e do ano seguinte ('shift').
#Por isto, também, a primeira linha de shift é NaN.

df_jj_agg['shift'] = df_jj_agg.shift(1)
```

```python
#Cria uma nova coluna com a diferença de rendimento com a subtração das colunas 'JohnsonJohnson' e 'shift'.
#Isto equivale à valoração do rendimento de um ano pra outro.

df_jj_agg['profit'] = df_jj_agg['JohnsonJohnson'] - df_jj_agg['shift']
df_jj_agg
```

```python
#2. Calcular a valorização do rendimento de um ano para outro

#Apenas retirando a coluna 'shift' para limpar um pouco o dataframe

df_jj_agg = df_jj_agg.drop(['shift'], axis=1)
df_jj_agg
```

```python
#3. Informar o ano de maior rentabilidade

#Com base no maior valor presente na coluna 'JohnsonJohnson', seleciona o ID. 
#Como o ID é correspondente ao ano, logo retorna o ano com maior rentabilidade.

df_jj_agg['JohnsonJohnson'].idxmax()                                    
```

```python
### Q4 - Qual percentual das pessoas apresentaram uma melhora de condições de moradia no final do estudo?
```

```python
#Verificar quem subiu 1 ou 2 níveis no indicador y da primeira para a quarta medição.

#Filtrar os dados da primeira medição (time = 0)
df_time_1 = df_housing.loc[df_housing['time'] == 0]

#Filtrar os dados da última medição (time = 24)
df_time_2 = df_housing.loc[df_housing['time'] == 24]
```

```python
#Unir os dois dataframes resultantes para poder calcular o percentual
df_merge = pd.merge(df_time_1, df_time_2, how = 'left', on = 'id')
```

```python
#Criar coluna como passo intermediário para verificar quem subiu de nível de moradia. 
# Pelo resultado da subtração é possível saber quem passou de 0.0 para 1.0 ou 2.0, e quem passou de 1.0 para 2.0.
df_merge['melhoria_moradia'] = df_merge['y_x'] - df_merge['y_y']
```

```python
#Identificar e excluir linhas onde as informações sobre as variáveis referentes ao status da moradia sejam nulas/NaN, pois não é possível identificar se houve melhora ou não devido a medição não ter sido realizada
df_merge = df_merge.dropna()
df_merge = df_merge.reset_index(drop = True)
```

```python
#Identificar e excluir casos onde as pessoas se mantiveram no mesmo status ou apresentaram piora na condição de moradia
sem_melhoria = df_merge[df_merge['melhoria_moradia'] >= 0.0].index
df_merge.drop(sem_melhoria, inplace=True)
```

```python
#Calcular o percentual: qtd de pessoas que apresentaram melhora/qtd total de pessoas
percentual_melhoria = (df_merge.shape[0]/df_time_1.shape[0]) * 100
```

```python
print(f'Cerca de {round(percentual_melhoria, 2)}% das pessoas acompanhadas apresentou uma melhora nas condições de moradia no final do estudo.')
```

**PASSO 1 DA RESOLUÇÃO DO DESAFIO!**

**1. Crie uma função (freq_suicide) que ao informar o sexo e a faixa etária de um suícidio é retornado as probabilidades do método utilizado.**

```python
def freq_suicide(dframe,sex,agegroup):
  dframe_suicide = dframe.groupby(['sex','age.group','method']).sum()
  dframe_suicide = dframe_suicide.reset_index()
  #dframe_suicide = dframe_suicide.drop(['Freq','age'], axis=1)

  df_suicide_freq = dframe_suicide[(dframe_suicide['sex'] == sex) & (dframe_suicide['age.group'] == agegroup)]
  df_suicide_freq['proportion'] = df_suicide_freq['Freq']/df_suicide_freq['Freq'].sum()

  return df_suicide_freq.drop(['Freq','age'], axis=1)
```
