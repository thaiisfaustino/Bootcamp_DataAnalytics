#####**Relação entre *lunch* e *test preparation course***

````python
# agrupar os dados
relacoes_quali = dados.drop(['math score','reading score', 'writing score', 'general mean', 'gender', 'parental level of education', 'race/ethnicity'], axis = 1)
relacoes_quali = relacoes_quali.groupby(
    ['lunch','test preparation course'],
    as_index = False,
    ).value_counts()
relacoes_quali.head()
````

````python
# Calcula a proporção de cada valor em relação ao total para cada grupo
total = relacoes_quali.groupby('lunch')['count'].transform('sum')
relacoes_quali['relative frequency'] = relacoes_quali['count'] / total
relacoes_quali.head(10)
````

````python
# plotar o gráfico usando o pacote plotly.express
fig = px.bar(relacoes_quali, x='lunch', y='relative frequency', color='test preparation course', 
             barmode='relative',
             category_orders={'test preparation course': ['sim', 'não'], 'lunch': ['De graça/ Taxa reduzida', 'Padrão']})
             

fig.show()
````

De acordo com a interpretação desse gráfico, quando a variável do eixo X apresenta uma composição muito semelhante em suas categorias, não podemos afirmar que há uma relação entre as variáveis.

### **Medidas de dispersão**

````python
amplitudeMath = dados['math score'].max() - dados['math score'].min()
amplitudeReading = dados['reading score'].max() - dados['reading score'].min()
amplitudeWriting = dados['writing score'].max() - dados['writing score'].min()

desvioMA_Writing = dados['writing score'].mad()
desvioMA_Reading = dados['reading score'].mad()
desvioMA_Math = dados['math score'].mad()

varianciaMath = dados['math score'].var()
varianciaReading = dados['reading score'].var()
varianciaWriting = dados['writing score'].var()

desviopadraoMath = dados['math score'].std()
desviopadraoReading = dados['reading score'].std()
desviopadraoWriting = dados['writing score'].std()
````

````python
medidas_dispersao = pd.DataFrame(
    {'Course': ['math', 'writing', 'reading'], 
     'Amplitude': [amplitudeMath, amplitudeWriting, amplitudeReading],
     'Desvio Médio Absoluto': [desvioMA_Math, desvioMA_Writing, desvioMA_Reading],
     'Variância': [varianciaMath, varianciaWriting, varianciaReading],
     'Desvio Padrão': [desviopadraoMath, desviopadraoWriting, desvioMA_Reading]
     },
     
)

medidas_dispersao
````

A disciplina com maior dispersão entre os dados é a disciplina de Escrita, seguida da disciplina de Matemática e Leitura.

### <font color=#334195>**2.**</font> Você notou diferenças nítidas pré e pós pandemia? Como você poderia provar estatisticamente que essas diferenças são significativas? **(pergunta extra)**

Após realizar a análise descritiva e identificar as distribuições dos dados, foi observada uma alteração no perfil dos clientes. No entanto, para comprovar estatisticamente essas diferenças, foram realizados testes específicos. O teste de Mann-Whitney foi utilizado para as variáveis quantitativas, enquanto o teste Qui-quadrado foi utilizado para as variáveis qualitativas. Os resultados indicam evidência estatística das diferenças no perfil do consumidor de produtos de beleza para cada uma das variáveis analisadas.

#### **Idade**

- **Hipótese nula** $H0$: a distribuição de idade dos consumidores antes e após a pandemia é a mesma. 
- **Hipótese alternativa** $Ha$: a distribuição de idade dos consumidores antes e após a pandemia é diferente.

````python
#Criando dataframes para as idades pré e pós-pandemia

idade_pre = dados_pre_pandemia['idade']
idade_pos = dados_pos_pandemia['idade']
````

````python
#Definindo o alpha ou nível de significância

alpha = 0.05
````

````python
#Teste de normalidade

teste1, p1 = shapiro(idade_pre)
print('Estatística do teste:', teste1)
print('p-valor:', p1)

if p1 < alpha:
    print("\nNão segue uma distribuição normal.")
else:
    print("\nSegue uma distribuição normal.")

teste2, p2 = shapiro(idade_pos)
print('\nEstatística do teste:', teste2)
print('p-valor:', p2)

if p2 < alpha:
    print("\nNão segue uma distribuição normal.")
else:
    print("\nSegue uma distribuição normal.")
````

````python
#Teste de igualdade de variâncias

teste3, p3 = bartlett(idade_pre, idade_pos)
print('Estatística do teste:', teste3)
print('p-value:', p3)

if p3 > alpha:
    print("\nA variância das amostras são iguais.")
else:
    print("\nA variância das amostras não são iguais.")
````

````python
#Teste Mann-Whitney

estatistica, p_valor = mannwhitneyu(idade_pre, idade_pos)

print("Estatística do teste:", estatistica)
print("p-valor: ", p_valor)

if p_valor < alpha:
    print("\nRejeitar hipótese nula. Há evidência estatística de que a média de idade mudou antes e depois da pandemia.")
else:
    print("\nAceitar a hipótese nula. Não há evidência estatística de que a média de idade mudou antes e depois da pandemia.")
````

````python
#Teste Qui-quadrado

estatistica2, p_valor2, graus_liberdade2, tabela_freq_esp2 = chi2_contingency(cont_table)

print('Estatística do teste: {}\n'.format(estatistica2) + 
      'p-valor: {}\n'.format(p_valor2) +
      'Graus de liberdade: {}\n\n'.format(graus_liberdade2) +
      'Tabela de Freq: \n{}'.format(tabela_freq_esp2))

if p_valor2 < alpha:
  print("\nRejeita-se a hipótese nula. Há evidência estatística para afirmar que a média de consumo de produtos antes da pandemia é diferente da média de consumo de produtos depois da pandemia.")
else:
  print("\nAceita-se a hipótese nula. Não há evidência estatística para afirmar que a média de consumo de produtos antes da pandemia é diferente da média de consumo de produtos depois da pandemia.")
````

A base informada é suficiente? E com uma margem de erro de **10%**?

````python
#Função para calcular margem de erro

def tamanho_amostra(N, e, confianca=0.95, p=0.5):
    z = stats.norm.ppf(1 - (1 - confianca) / 2)
    q = 1 - p
    n = (z ** 2 * p * q * N) / ((z ** 2 * p * q) + (e ** 2 * (N - 1)))
    return math.ceil(n)
````

````python
#Calcular margem de erro de 2%

tamanho_amostra(N=100000, e=0.02, confianca=0.95)
````

## <font color=#647AA3>**Desafio 2:** Determinando o Estoque de Segurança.</font>

Suponha agora que você trabalha em uma empresa de produtos perecíveis, como Danone ou Nestlé. Para a empresa, é muito importante que a sua **política de estoque seja eficiente**. É necessário manter um estoque de segurança em todos os produtos **(acima da previsão de vendas)**, pois, caso contrário, as pessoas podem não encontrar os produtos no supermercado. Mas, ao mesmo tempo, se o estoque estiver em excesso, isso pode representar um grande custo. Os produtos podem não ser comprados, e consequentemente, jogados fora, o que na indústria se chama ***desova***.

Para achar o equilíbrio perfeito, o gerente de vendas pediu a ajuda a você, do time de análise de dados, e ele gostaria de saber qual o estoque de segurança ideal.

Para responder esse desafio, antes vamos abordar alguns conceitos importantes na indústria:


1. **Estoque de segurança:** é uma quantidade de estoques que se tem, mas que não se deseja usar. É como um seguro do carro: é bom ter um, mas você não quer precisar usá-lo. Isto porque ele é calculado apenas para suprir uma segurança em casos de variações inesperadas, algo que você não pode prever ou controlar: problemas no fornecedor, atrasos na entrega, uma demanda que não foi prevista.
2. **Demanda:** se a demanda é bem estável e conhecida com antecedência,
então temos pouca variabilidade a cada mês e não precisamos nos proteger
muito contra essas variações (pois sabemos que elas não ocorrem); por
outro lado, se seu produto tem uma variabilidade nas vendas muito grande,
então precisaremos de estoque de segurança maior. Isto tudo é medido
matematicamente através do desvio padrão da demanda, que neste caso é
calculado como o desvio padrão da previsão da demanda. Um bom sistema
de previsões é capaz de oferecer este número, ou ele pode ser estimado de
maneiras mais simples, mas menos precisas.
3. **Lead time** (tempo de entrega) do produto: se o tempo de entrega é elevado
e sua variabilidade é alta (se uma entrega é feita em 5 dias, outra em 8 dias,
outra em 2 dias), então é preciso ter uma segurança frente à este tempo
média de entrega de 5 dias, pois algumas vezes ela chega a demorar 8 dias.
Mas não queremos nos planejar sempre para receber apenas depois de 8
dias, pois isto acarretaria custos muito altos, então o estoque de segurança
utiliza a estatística para auxiliar nessa tarefa.
4. **Nível de serviço desejado:** nem todos os produtos merecem a mesma
atenção e o mesmo cuidado; alguns produtos são críticos, mais importantes
ou mais atrativos, e por isso merecem estar sempre presentes, enquanto
em outros produtos podemos nos dar ao luxo de não tê-lo em estoque
sempre. Matematicamente, isto é modelado através do nível de serviço
desejado: quanto maior o nível de serviço (um número percentual de 0 a
100), maior será o estoque de segurança pois queremos mais garantias que
o produto estará sempre disponível. O nível de serviço depende de cada
setor: palitos de fósforo num supermercado não devem ter nível de serviço
muito alto, enquanto antibióticos numa farmácia hospitalar devem ter nível
de serviço altíssimo. O nível de serviço indica o quanto queremos estar
seguros frente às variabilidades que ocorrem, em outras palavras, frente
aos desvios padrões da demanda e do lead time.
5. **A Fórmula Mãe do Estoque de segurança:**
A maioria das indústrias calcula o estoque de segurança tomando como
base a distribuição normal. Assumindo que suas vendas seguem essa
distribuição, ela irá determinar qual a probabilidade de se vender certo
produto. image.png

Por exemplo, se você vende na média 1000 unidades de produto por semana, a distribuição mostra que você tem uma alta probabilidade de vender em torno de 1000 todas as semanas, e uma baixa probabilidade de esse número ser 20 ou 200000 de produtos vendidos. Assim, você pode utilizar a distribuição normal, para entender qual o seu nível de serviço desejado. Ou seja, se você deseja um nível de serviço de 90%, a curva normal irá te fornecer o multiplicador desejado (variações da
média) para você alcançar 90% da demanda.
Portanto, matematicamente, podemos determinar o estoque de segurança utilizando os conceitos abordados em estatística, como a tabela da distribuição normal, utilizando a fórmula abaixo.

**Estoque de Segurança** 

# $$𝑍 * \sqrt{σ_𝑑^2𝑡 + σ_𝑡^2𝑑^2}$$ 

sendo:
- $z$ = o valor da tabela da distribuição normal, que indica a quantos desvios padrões ao redor da média temos que tomar para cobrir a proporção da área sob a curva normal /nível de serviço desejado.
- $d$ = demanda média passada ou estimada
- $t$ = lead time médio.
- $σ_d$ = desvio padrão / variação da demanda
- $σ_t$ = desvio padrão / variação do lead time

Assim se o tempo de entrega tiver pouca variabilidade a fórmula pode ser aproximada por:

**Estoque de Segurança**

#$$𝑍 * \sqrt{σ_𝑑^2𝑡} $$

e a intuição da fórmula pode ser vista como : O quanto de variação da
média da demanda eu estou disposto a aceitar além do natural (z-score do
nível de serviço) * desvio médio da demanda que já ocorre naturalmente *
tempo de entrega.

Matematicamente o valor obtido pela fórmula deve ser sempre
arredondado para cima, para garantir que estamos cobertos contra
variações indesejadas.

````python
# A realização do teste de Kruskal-Wallis nos mostro que pelo menos uma fórmula apresenta diferença significativa em relação as outras, porém não indicou qual.
# Para identificar quais foram os grupos que apresentaram diferenças significativas entre si é necessário aplicar um teste post-hoc.
# O teste de comparações múltiplas escolhido foi o de Dunn.
````

````python
p_valor = posthoc_dunn([formula_1_sol, formula_2_sol, formula_3_sol])
p_valor
````

````python
# A saída do teste de Dunn é uma tabela que compara os grupos de dois em dois, e retorna os p_valor. 
# Quando o p_valor for menor ou igual ao nível de significância empregado no teste, pode-se rejeitar a hipóteses nula, indicando que existe diferença entre os grupos em questão.
# No caso, observa-se que todos os grupos se diferem entre si, o que nos leva a concluir que as 3 fórmulas apresentam diferenças significativas de performance quando expostas ao sol.
````

````python
# Partindo da confirmação que as 3 fórmulas apresentam diferenças entre sim, avaliou-se o resultado do experimento com as participantes para indicar o batom que apresentasse maior durabilidade ao ser exposto ao sol.
# Amostra - Com filtro para o campo exposicao_sol = 1

media_form_1 = formula_1_sol.mean()
print(f'A durabilidade média da fórmula 1 para participantes que se expuseram ao sol foi de {media_form_1.round(2)}')

media_form_2 = formula_2_sol.mean()
print(f'A durabilidade média da fórmula 2 para participantes que se expuseram ao sol foi de {media_form_2.round(2)}')

media_form_3 = formula_3_sol.mean()
print(f'A durabilidade média da fórmula 3 para participantes que se expuseram ao sol foi de {media_form_3.round(2)}')
````

Imagine que voce trabalha em uma empresa do setor alimentício, e o preço de

**7,85** reais e desvio padrão igual a **3,32** reais. Suspeita-se que devido ao verão, em
determinadas localidades de praia o preco do picolé esteja mais caro.

Para verificar se esse efeito é estatisticamente significativo, contactaram voce, analista
de dados.

Suponha que para realizar essa análise voce tenha dados de **45** lojas em cidades do
litoral, escolhidas aleatoriamente. E constatou que o preço médio do picolé nessas
cidades é de **11,52** reais.


### <font color=#334195>**1.**</font> Voce pode concluir que existe diferenca de precos estatisticamente relevante com 5% de significancia?


````python
# Importando bibliotecas

import pandas as pd
import numpy as np
from scipy.stats import norm
````

````python
# Dados do problema

media = 7.85
desvio_padrao = 3.32
n = 45
media_amostral = 11.52
````

````python
# Passo 1 - Definir hipóteses nula e alternativa

#H0: Média = 7,85 , ou seja, não existe diferença de preço entre localidades
#H1: Média != 7,85 , ou seja, existe diferença de preço entre localidades
````

````python
# Passo 2 - Escolher distribuição amostral

# n >= 30
# desvio padrão populacional é conhecido
# Utilizaremos a distribuição normal
````

````python
# Passo 3 - Fixação da significância do teste

# Significância de 5%
# Teste bicaudal, pois H1 definida é do tipo '!='

significancia = 0.05

confianca = 1 - significancia

probabilidade = (0.5 + (confianca / 2))

z_alpha_2 = norm.ppf(probabilidade)
z_alpha_2.round(2)
````

````python
# Passo 4 - Cálculo da estatística teste

z = (media_amostral - media) / (desvio_padrao / np.sqrt(n))
z.round(2)
````

````python
# Passo 5  - Aceitação ou Rejeição da hipótese nula

# Critério do valor crítico para teste bicaudal:
# Rejeitar H0 se z <= - z_alpha_2 ou se z >= z_aplha_2
````

````python
z <= -z_alpha_2
````

````python
z >= z_alpha_2
````

````python
# Conclusão

print('Conclusão: A partir da análise da estatística teste, ao nível de 5% de significância, rejeita-se a hipótese nula \n de que a média de preço é igual a R$ 7,85, confirmando a existência de diferença de preços \n em determinadas localidades.')
````

### <font color=#334195>**2.**</font> Qual o valor da variável teste, o intervalo de confianca e o p-valor?

````python
# Intervalo de confiança

confianca = 1 - significancia

desvio_padrao = 3.32
n = 45
raiz_n = np.sqrt(n)
sigma = desvio_padrao / raiz_n
e = z*sigma
````

````python
i= intervalo = (
  media_amostral - e,
  media_amostral + e  
)
intervalo
````

````python
print(f'O nível de confiança do teste realizado é de {confianca * 100}%, \n com um intervalo de confiança igual a {intervalo}.')
````

````python
# Valor da variável teste

print(f'O valor da variável teste é de {z.round(2)}, calculado no item anterior.')
````

````python
# P-valor

p_valor = 2 * (1 - norm.cdf(z))
p_valor
````

````python
p_valor <= significancia
````

````python
print(f'Conclusão: Ao analisar o p-valor obtido de {p_valor}, podemos reafirmar a conclusão de rejeição da hipótese nula,\numa vez que ele é menor do que a significância utilizada no teste.')
````

````python
#Definindo a variável 'price' como target - y

y = df_house_price['price']
````

````python
#Função criada para dividir o dataframe em teste e treino

def train_test(x_variable, y_variable):
    X_train, X_test, y_train, y_test = train_test_split(x_variable, y_variable, test_size=0.3, random_state=2811)
    
    return X_train.values.reshape(-1, 1), X_test.values.reshape(-1, 1), y_train.values.reshape(-1, 1), y_test.values.reshape(-1, 1)
````

````python
#Função para calcular diferentes métricas para avaliar o modelo. 
#Usando variáveis x e y de treino.

# model.score(X, y): R² - Coeficiente de Determinação em teste, o quanto a reta se ajusta ao modelo.
# coef_: Coeficientes das features no modelo
# intercepto: valor em que a reta ajustada cruza o eixo y

def model_training(X_training, y_training, bool_print = True):
    model = LinearRegression().fit(X_training, y_training)
   
    try:
        intercept_ = model.intercept_
    except:
        intercept_ = 0

    try:
        coef_ = model.coef_
    except:
        coef_ = 0 
    
    
    if bool_print:
        print("• R²: ", model.score(X_training, y_training))
        print("• Coeficientes: ", coef_)
        print("• Intercepto: ", intercept_)
        
        
    return model.score(X_training, y_training), intercept_, coef_, model
````

````python
#Função para calcular métricas entre y de treino e y previsto

# Erro Quadrático Médio: é entre y previsto e y de treino, pode ser usado como medida da qualidade de um estimador
# Raiz do EQM: representa a diferença entre valores previsto e observados no modelo
# model.score(X, y): R² - Coeficiente de Determinação em teste, o quanto a reta se ajusta ao modelo.


def model_testing(y_testing, y_predict, bool_print = True):
    model = LinearRegression().fit(y_testing, y_predict)
    
    EQM_2 = metrics.mean_squared_error(y_testing, y_predict)
    REQM_2 = np.sqrt(metrics.mean_squared_error(y_testing, y_predict))
    try:
        coef_ = model.coef_
    except:
        coef_ = 0 
    
    
    if bool_print:
        print("• Erro Quadrático Médio: ", EQM_2)
        print("• Raiz do Erro Quadrático Médio: ", REQM_2)
        print("• R²: ", model.score(y_testing, y_predict))
        
        
    return EQM_2, REQM_2, model.score(y_testing, y_predict)
````

````python
# Verificar a existência de valores ausentes (NaN)
imovel.isnull().sum()
````

````python
# Armazenar a variavel dependente (Y)
y = imovel['price_log']
````

````python
# Armazenar as variáveis explicativas
X = imovel[['sqft_living', 'grade', 'sqft_above', 'bathrooms', 'view']]
````

Criar os datasets de treino e de teste:

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2811)
````

````python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y, test_size=0.3, random_state=2811)
````

````python
modelo_2.fit(X2_train, y2_train)
````

Comparar o coeficiente de determinação (R²) do novo modelo estimado com o resultado do modelo anterior:

Métricas de regressão (avaliar a qualidade do modelo):

````python
EQM = metrics.mean_squared_error(y_test, y_previsto).round(2)
REQM = np.sqrt(metrics.mean_squared_error(y_test, y_previsto)).round(2)
R2 = metrics.r2_score(y_test, y_previsto).round(2)

pd.DataFrame([EQM, REQM, R2], ['EQM', 'REQM', 'R²'], columns=['Métricas'])
````

````python
EQM = metrics.mean_squared_error(y2_test, y2_previsto).round(2)
REQM = np.sqrt(metrics.mean_squared_error(y2_test, y2_previsto)).round(2)
R2 = metrics.r2_score(y2_test, y2_previsto).round(2)

pd.DataFrame([EQM, REQM, R2], ['EQM', 'REQM', 'R²'], columns=['Métricas'])
````

Interpretando os resultados:

- EQM (Erro Quadrático Médio): Indica a capacidade do modelo de prever com precisão o preço dos imóveis. Um EQM baixo indica que o modelo tem uma boa capacidade de realizar essa previsão. 

- REQM (Raiz Quadrada do Erro Quadrático Médio): Fornece uma medida da diferença média entre os valores preditos pelo modelo e os valores observados, neste caso, em unidades de preço. 

- R² (Coeficiente de Determinação): Varia de 0 a 1, sendo que um R² próximo de 1 indica que o modelo explica uma grande parte da variação, nesse caso, nos preços dos imóveis. Quanto maior o valor de R², melhor o modelo se ajusta aos valores reais. Aqui, os valores previstos e de teste de ambos os modelos possuem 59% de correlação.

Conclusão: Baseando-se apenas nas métricas de regressão apresentadas acima, os dois modelos que estão sendo comparados não apresentaram diferença em relação a sua qualidade, portanto foi mantido o Modelo 1.

Obs: A escolha de um modelo não baseia-se apenas nas três métricas calculadas. Em casos como esse poderia ser realizado uma análise mais detalhada de ambos, levando também em consideração questões como o desempenho de cada modelo frente a dados e não vistos anteriormentes e a relevância das variáveis explicativas que estão sendo incluídas, por exemplo.
